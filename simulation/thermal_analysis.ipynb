# Thermal Analysis
This notebook recreates thermal plots from CSV data.

```python
# Thermal Analysis
This notebook recreates thermal conductivity and temperature plots from CSV data.

import pandas as pd
import matplotlib.pyplot as plt

thermal = pd.read_csv('../results/thermal_conductivity.csv')
thermal.plot(x='Fluid', y='Thermal_Conductivity_W_per_mK', kind='bar', legend=False)
plt.ylabel('W/m·K')
plt.title('Thermal Conductivity by Fluid Variant')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

cc = pd.read_csv('../results/cutting_temperatures_coconut.csv')
cc.plot(x='Time_min', y=[c for c in cc.columns if c!='Time_min'])
plt.ylabel('°C')
plt.title('Cutting Temperature vs Time (Coconut Variants)')
plt.tight_layout()
plt.show()

sf = pd.read_csv('../results/cutting_temperatures_sunflower.csv')
sf.plot(x='Time_min', y=[c for c in sf.columns if c!='Time_min'])
plt.ylabel('°C')
plt.title('Cutting Temperature vs Time (Sunflower Variants)')
plt.tight_layout()
plt.show()
```
