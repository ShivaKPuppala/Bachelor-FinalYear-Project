# Wear Rate Visualization
Bar chart from the tool wear dataset.

```python
# Wear Rate Visualization
import pandas as pd
import matplotlib.pyplot as plt

wear = pd.read_csv('../results/tool_wear_data.csv')
wear.plot(x='Fluid', y='Tool_Wear_micrometers', kind='bar', legend=False)
plt.ylabel('µm')
plt.title('Tool Wear by Fluid Variant')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
